```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.constants as ct
from scipy import integrate
plt.style.use('default')
from matplotlib import lines
from matplotlib import collections  as mc
```

```python
#Fils
def V2(x, y):
    r0=1
    L=1
    r=np.sqrt(x**2+y**2)
    V=L/(2*np.pi*ct.epsilon_0)*np.log(r0/r)

    return V
def V1(x, y):
    r0=1
    L=-1
    r=np.sqrt(x**2+y**2)
    V=L/(2*np.pi*ct.epsilon_0)*np.log(r0/r)

    return V
nx, ny = 64, 64
a=4
b=8
x = np.linspace(-b, b, nx)
y = np.linspace(-a, a, ny)
X, Y = np.meshgrid(x, y)
Z1=V1(X+d,Y)
d=1
Z2=V2(X-d,Y)
#Grfics

plt.contour(X,Y,Z1+Z2,20,colors='black')
plt.show()
```

```python
#Condensador
h=1
dd=10
L=1
k=1/(4*np.pi*ct.epsilon_0)
def V1(x, y):
    V=integrate.quad(lambda d: k*L/(np.sqrt((x-d)**2+(y-h)**2)), 0, dd)
    return V[0]
def V2(x, y):
    V=integrate.quad(lambda d: k*-L/(np.sqrt((x-d)**2+(y+h)**2)), 0, dd)
    return V[0]

nx=100
ny=nx
a=2*dd
# a=2
x = np.linspace(-a, a+dd, nx)
y = np.linspace(-a, a+dd, ny)
X, Y = np.meshgrid(x, y)

Vv1=np.vectorize(V1)
Vv2=np.vectorize(V2)

Z1=Vv1(X,Y)
Z2=Vv2(X,Y)
fig, ax = plt.subplots(figsize=(12, 6))


ax.contour(X,Y,Z1+Z2,20,colors='black')
aa=3
e=6
plt.xlim([-aa,dd+aa])
plt.ylim([-e,e])
# lines=[[(0,h),(dd,h)],(0,-h),(dd,-h)]
lines = [[(0, 1), (1, 1)], [(2, 3), (3, 3)], [(1, 2), (1, 3)]]
x=[0,dd]
y=[h,h]
y2=[-h,h]

ax.plot(x,y)
# Z1=V(X,Y)
# d=1
# Z2=V(X,Y-1)
#Grfics
```
