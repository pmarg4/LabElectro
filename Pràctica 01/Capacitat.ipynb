```python
p=pd.read_csv("Capacitat.csv",sep=';',decimal=',')
# ,na_filter=True,skip_blank_lines=True
# p3=p3.dropna()
# p3=p3.drop([3])
# P['dl'] = p['dl'].astype(float64)

p.head()
p["dl"]=p["dl"]/100
p["ri"]=p["ri"]/100
udl=1/1000
uri=1/1000
uV=0.01
l=len(p["ri"])-1
ui1=np.empty(l)
ui2=np.empty(l)
ui3=np.empty(l)
nu=np.empty(l)
i=1
while i<l:
    ui1[i]=p.at[i,"dl"]/p.at[i,"ri"]
    i=i+1
i=1
    
while i<l:
    ui2[i]=(p.at[i,"V1"]-p.at[i,"V2"])/p.at[i,"ri"]
    i=i+1
i=1

while i<l:
    ui3[i]=(p.at[i,"V1"]-p.at[i,"V2"])*p.at[i,"dl"]/p.at[i,"ri"]**2
    i=i+1
i=1
    
while i<l:
    nu[i]=uV**2*ui1[i]**2+udl**2*ui2[i]**2+uri**2*ui3[i]**2
    i=i+1
suma=0    
i=1

while i<l:
    suma=suma+nu[i]
    i=i+1

u=np.sqrt(suma)
uqz=suma
print(suma)
```
